```python
well_drop_dup = wellnes_col.drop_duplicates()
well_drop_dup
```

```python
wellnes_col
```

```python
wellness_result_grp = wellnes_col.groupby("DocumentNo")["Credit Notes total"].sum().reset_index()
wellness_result_grp
```

```python
wellness_result_grp2 = well_drop_dup.groupby("DocumentNo")["Credit Notes total"].sum().reset_index()
wellness_result_grp2
```

```python
wellness_total_result = pd.merge(left = wellness_result_grp2, right= wellnessjuly_copy, how= "inner",left_on= "DocumentNo",right_on="DocumentNo", indicator= True)
wellness_total_result
```

```python
Wellness_july_data = wellness_total_result[wellness_total_result["_merge"]== "both"]
Wellness_july_data
```

```python
with pd.ExcelWriter(r"C:\Users\user\Downloads\wellnessData.xlsx", engine ="openpyxl") as dt:
    wellness_result_grp2.to_excel(dt,sheet_name="wellness_result_grp2",index = False)
    wellness_total_result.to_excel(dt,sheet_name= "wellness_total_result",index = False)
    Wellness_july_data.to_excel(dt,sheet_name="wellness_july_data",index=False)
    
```
