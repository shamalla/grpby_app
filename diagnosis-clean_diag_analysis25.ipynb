```python
#dg will be the dhortform of diagnosis
#cd is for the code column
dg = pd.read_excel(r"C:\Users\user\Downloads\Data Sample.xlsx", sheet_name="edited_data")
cd = pd.read_excel(r"C:\Users\user\Downloads\Data Sample.xlsx", sheet_name="testcodes")
```

```python
dg_desc = dg.copy()
dig_col = ['diagnosis1', 'diagnosis2','diagnosis3', 'diagnosis4', 'diagnosis5', 'diagnosis6', 'diagnosis7','diagnosis8', 'diagnosis9', 'diagnosis1a']
for col in dig_col:
    dg_desc = dg_desc.merge(cd,left_on=col,right_on='ICD10_Code',how="left")
    dg_desc=dg_desc.rename(columns={'WHO Full Description':f"{col}_desc"})
    dg_desc=dg_desc.drop(columns=['ICD10_Code'])
```

```python
dg_desc
```

```python
dg_desc.columns
```

```python
#We are combining the text and not including nulls in our combined text
description = ['diagnosis1_desc', 'diagnosis2_desc', 'diagnosis3_desc','diagnosis4_desc', 'diagnosis5_desc', 'diagnosis6_desc','diagnosis7_desc', 'diagnosis8_desc', 'diagnosis9_desc','diagnosis1a_desc']
dg_desc["DSCPT"] = dg_desc[description].agg(lambda x:",".join([str(i) for i in x if pd.notnull(i)]), axis=1)
```

```python
dg_desc
```

```python
diagnosis = ['diagnosis1', 'diagnosis2','diagnosis3', 'diagnosis4', 'diagnosis5', 'diagnosis6', 'diagnosis7','diagnosis8', 'diagnosis9', 'diagnosis1a']
dg_desc["Diagnosis"] = dg_desc[diagnosis].agg(lambda x: "-".join([str(i) for i in x if pd.notnull(i)]), axis=1)
```

```python
dg_desc
```

```python
diagnosis = ['diagnosis1', 'diagnosis2','diagnosis3', 'diagnosis4', 'diagnosis5', 'diagnosis6', 'diagnosis7','diagnosis8', 'diagnosis9', 'diagnosis1a']
dg_desc["Orig_diag"] = dg_desc[diagnosis].agg(lambda x: ",".join([str(i) for i in x if pd.notnull(i)]), axis=1)
```

```python
dg_desc
```

```python
dg_desc["final_encounter_nr"] = dg_desc["final_encounter_nr"].astype("Int64").astype(str)
```

```python
dg_desc
```

```python
dg_desc.columns
```

```python
dg_desc['ChargeDate'] = pd.to_datetime(dg_desc['ChargeDate'], errors=("coerce"))
```

```python
#We want to replace october, november and december 2024 to january 2025
month_to_change = [10,11,12]
mask = (dg_desc['ChargeDate'].dt.year == 2024)&(dg_desc['ChargeDate'].dt.month.isin(month_to_change))
dg_desc.loc[mask,'ChargeDate'] = dg_desc.loc[mask,'ChargeDate'].apply(lambda d: d.replace(year = 2025, month = 1))
```

```python
dg_desc["charge_m_y"] = dg_desc['ChargeDate'].dt.strftime("%B-%Y")
```
