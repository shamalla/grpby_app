```python
#So the above can also be written in a pivot table format so that if i do not want each visit id to be reflected and i want to know just the amount we made in each location per month
#So we then use the pivot table
Visit_amount_per_location_month = visit_id.pivot_table(index=["date_yr"],columns="Location",values="TotalAmount",aggfunc=["sum","mean"])
#Visit_amount_per_location_month
```

```python
Visit_amount_per_location_month_COUNT =(
    visit_id.pivot_table(
        index=["date_yr"],
        columns="Location",
        values="VisitID",
        aggfunc=pd.Series.nunique
    ).reset_index()
)
Visit_amount_per_location_month_COUNT
```

```python
Visit_amount_per_location_month_COUNT.to_excel(r"C:\Users\user\Downloads\visit_unique_count.xlsx")
```

```python
visit_per_servise_per_month = visit_id.pivot_table(index=["date_yr","Location"], columns="Service",values="TotalAmount",aggfunc=["sum","mean"])
#visit_per_servise_per_month
```

```python
visit_per_servise_per_month.fillna(0, inplace=True)
#visit_per_servise_per_month
```

```python
visit_per_servise_per_month_location = visit_id.pivot_table(index=["date_yr","Service"], columns="Location",values="TotalAmount",aggfunc=["sum","mean"])
#visit_per_servise_per_month_location
```

```python
visit_per_servise_per_month_location.fillna(0, inplace=True)
#visit_per_servise_per_month_location
```

```python
Visit_amount_per_location_month_unique = (
    visit_id.pivot_table(
        index=["date_yr","Service"],
        columns="Location",
        values='VisitID',
        aggfunc=pd.Series.nunique
    ).reset_index()
)
Visit_amount_per_location_month_unique
```

```python
Amount_service_month = (
    visit_id.pivot_table(
        index=["Service"],
        columns="date_yr",
        values="TotalAmount",
        aggfunc=["sum"]
    ).reset_index()
)
#Amount_service_month
```
